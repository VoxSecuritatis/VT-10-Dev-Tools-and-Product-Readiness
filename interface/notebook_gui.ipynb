# Multi-Agent Startup Pitch Planner -- Notebook GUI

Enter a startup domain (for example `fintech` or `healthtech`) and run the workflow.
The Research Agent, Funding Advisor, and Pitch Coach run in sequence, with a feedback
loop back to an earlier agent if the Pitch Coach decides more detail is needed.

This notebook calls the same `workflow.graph.run_workflow()` entry point as
`interface/cli.py` -- no logic is duplicated between the two interfaces.

```python
import sys
from pathlib import Path

project_root = Path.cwd().parent if Path.cwd().name == "interface" else Path.cwd()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

import ipywidgets as widgets
from IPython.display import display

from workflow.graph import run_workflow
```

```python
domain_input = widgets.Text(
    value="",
    placeholder="e.g. fintech, healthtech, climate tech",
    description="Domain:",
    style={"description_width": "initial"},
)
run_button = widgets.Button(description="Generate Pitch Outline", button_style="warning")
output_area = widgets.Output()


def on_run_button_clicked(_button: widgets.Button) -> None:
    """Run the workflow for the entered domain and render the outline and trace summary."""
    output_area.clear_output()
    domain = domain_input.value.strip()
    with output_area:
        if not domain:
            print("[WARN] Enter a startup domain before running.")
            return
        print(f"Running workflow for domain: {domain} ...\n")
        final_state = run_workflow(domain)
        print("Pitch Deck Outline\n" + "=" * 18 + "\n")
        print(final_state["pitch_outline"])
        print(f"\n[INFO] revisions used: {final_state['revision_count']}")
        print(f"[INFO] trace log: logs/run_{final_state['run_id']}.jsonl")


run_button.on_click(on_run_button_clicked)
display(widgets.VBox([domain_input, run_button, output_area]))
```
